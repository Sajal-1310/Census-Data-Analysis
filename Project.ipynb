```python
df.groupby('State_name').Population.sum()
#The groupby() function is used to group data based on one or more columns,
#and the .sum() function is used to combine the values (specifically summing the numeric columns) for each group.
```

```python
df.groupby('State_name').Population.sum().sort_values()
#It allows you to sort the data in either ascending or descending order based on the values of the specified column
```

```python
df.groupby('State_name')['Population'].sum().plot(kind='bar', figsize=(10, 8), color='black')
plt.xlabel('State Name', fontsize=14, color='b')
plt.ylabel('Total Population', fontsize=14, color='b')
plt.title('Total Population by State', fontsize=16, color='green')
plt.grid(axis='y')
plt.show()
```

```python
df.groupby('State_name')['Male'].sum().plot(kind='bar', figsize=(10, 8), color='blue')
plt.xlabel('State Name', fontsize=14, color='b')
plt.ylabel('Male Population', fontsize=14, color='b')
plt.title('Male Population by State', fontsize=16, color='blue')
plt.grid(axis='y')
plt.show()
```

```python
df.groupby('State_name')['Female'].sum().plot(kind='bar', figsize=(10, 8), color='aqua')
plt.xlabel('State Name', fontsize=14, color='b')
plt.ylabel('Female Population', fontsize=14, color='b')
plt.title('Female Population by State', fontsize=16, color='blue')
plt.grid(axis='y')
plt.show()
```

```python
df.groupby('Literate')['Population'].sum().plot(kind='line', figsize=(10, 8), color='Black')
plt.xlabel('Literate', fontsize=14, color='b')
plt.ylabel('Population', fontsize=14, color='b')
plt.title('Population by Literacy', fontsize=16, color='green')
plt.grid(axis='y')
plt.show()
```

```python
df.groupby('Literate')['Female'].sum().plot(kind='line', figsize=(10, 8), color='red')
plt.xlabel('Literate', fontsize=14, color='b')
plt.ylabel('Female Population', fontsize=14, color='b')
plt.title('Female Population by Literacy', fontsize=16, color='green')
plt.grid(axis='y')
plt.show()
```

```python
df.groupby('Literate')['Male'].sum().plot(kind='line', figsize=(10, 8), color='blue')
plt.xlabel('Literate', fontsize=14, color='b')
plt.ylabel('Male Population', fontsize=14, color='b')
plt.title('Male Population by Literacy', fontsize=16, color='green')
plt.grid(axis='y')
plt.show()
```

```python
df.groupby('Workers')['Population'].sum().plot(kind='hist', figsize=(10, 8), color='Black')
plt.xlabel('Workers', fontsize=14, color='b')
plt.ylabel('Population', fontsize=14, color='b')
plt.title('Total population of workers', fontsize=16, color='green')
plt.grid(axis='y')
plt.show()
```

```python
df.groupby('Workers')['Male_Workers'].sum().plot(kind='hist', figsize=(10, 8), color='orange')
plt.xlabel('Workers', fontsize=14, color='b')
plt.ylabel('Male_Workers', fontsize=14, color='b')
plt.title('Population of Male workers', fontsize=16, color='green')
plt.grid(axis='y')
plt.show()
```

```python
df.groupby('Workers')['Female_Workers'].sum().plot(kind='hist', figsize=(10, 8), color='pink')
plt.xlabel('Workers', fontsize=14, color='b')
plt.ylabel('Female_Workers', fontsize=14, color='b')
plt.title('Population of Female workers', fontsize=16, color='green')
plt.grid(axis='y')
plt.show()
```

```python
sns.displot(df['Population'], kde=True)
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Distribution of Population')
plt.show()
```

```python
sns.displot(df['Literate'], kde=True)
plt.xlabel('Literate')
plt.ylabel('Density')
plt.title('Literacy of Population')
plt.show()
```

```python
sns.relplot(x='Population', y='Literate', data=df)
plt.title('Litracy of Population')
```

```python
sns.relplot(x='Age_Group_50', y='Literate', data=df)
plt.title('Litracy of age group 50')
```
